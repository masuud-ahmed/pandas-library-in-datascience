```python
so=df[df['country']=='Somalia']
```

```python
k=df[df['country']=='Kenya']
```

```python
k
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))

plt.style.use('seaborn-v0_8-whitegrid')

plt.plot(so.year,so.population,marker='s',label='Somalia')
plt.plot(k.year,k.population,marker='*',label='Kenya')

plt.title ('Somalia VS Kenya')
plt.ylabel('populations')
plt.xlabel('years')

plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.style.use('seaborn-v0_8-deep')

plt.subplot(2,1,1)
plt.plot(so.year,so.population,'r--',marker='s',label='Somalia')
plt.legend()

plt.subplot(2,1,2)
plt.plot(k.year,k.population,'g--',marker='*',label='Kenya')
plt.legend()
```
